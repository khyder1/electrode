```python
import matplotlib.pyplot as plt
from electrode import Brain, LIAFNeuron, Electrode, BasicCurrentSynapse
```

```python
rec = []
B = Brain()

B.add_neuron("A1", LIAFNeuron())
B.add_neuron("B", LIAFNeuron())
B.add_neuron("A2", LIAFNeuron())

B.add_synapse("A1", "B", BasicCurrentSynapse(current=3))
B.add_synapse("A2", "B", BasicCurrentSynapse(current=6))

E1 = B.add_electrode("A1")
E2 = B.add_electrode("A2")
F = B.add_electrode("B")

for i in range(5000):
    B.step()
    if i > 50:
        E1.pin(1.4)
    if i > 2000:
        E2.pin(9)
    rec.append((E1.measure(), F.measure()))
```

```python
plt.plot([r[0] for r in rec], label="A1")
plt.plot([r[1] for r in rec], label="B")
plt.legend()
```
