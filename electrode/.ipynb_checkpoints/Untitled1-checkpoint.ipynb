```python
from brian2 import *
import networkx as nx
import matplotlib.pyplot as plt
from __init__ import *
import pandas as pd
import ipywidgets as widgets
```

```python
connectome = nx.read_graphml('c_elegans_control.graphml')
duration = 200
rec = []
B = Brain()
ablateDDnodes = ['DD1','DD2','DD3','DD4','DD5','DD6']
ablateVDnodes = ['VD1', 'VD2', 'VD3', 'VD4', 'VD5', 'VD6', 'VD7', 'VD8', 'VD9', 'VD10', 'VD11', 'VD12', 'VD13']
```

```python
ablation = widgets.Dropdown(
    options=['Yes','No'],
    value='No',
    description='Ablation:',
    disabled=False,
)
ablation
```

```python
if ablation.value == 'Yes':
    ablate_nodes = widgets.Dropdown(
    options=['DD','VD'],
    value='VD',
    description='Ablated Nodes:',
    disabled=False,
)
ablate_nodes
```

```python
if ablate_nodes.value == 'DD':
    connectome.remove_nodes_from(ablateDDnodes)
if ablate_nodes.value == 'VD':
    connectome.remove_nodes_from(ablateVDnodes)

electrodes = []
for node in connectome.nodes():
    B.add_neuron(node, LIAFNeuron())
    electrodes.append(B.add_electrode(node))
electrode_names = dict(zip(list(connectome.nodes()),electrodes))

node_names = list(dict(connectome.nodes()).keys())
muscles = [i for i in node_names if i.startswith('M')]
dorsal_muscles = [i for i in node_names if i.startswith('MD')]
ventral_muscles = [i for i in node_names if i.startswith('MV')]

muscle_electrodes = []
for e in electrode_names:
    if e in muscles:
        muscle_electrodes.append(electrode_names[e])
muscle_electrodes_dict = dict(zip(list(muscles),muscle_electrodes))

dorsal_muscle_electrodes = []
for e in electrode_names:
    if e in dorsal_muscles:
        dorsal_muscle_electrodes.append(electrode_names[e])
dorsal_muscle_electrodes_dict = dict(zip(list(dorsal_muscles),dorsal_muscle_electrodes))

ventral_muscle_electrodes = []
for e in electrode_names:
    if e in ventral_muscles:
        ventral_muscle_electrodes.append(electrode_names[e])
ventral_muscle_electrodes_dict = dict(zip(list(ventral_muscles),ventral_muscle_electrodes))
 
for u,v,z in connectome.edges(data='weight', default = 0.1):
    B.add_synapse(u, v, BasicCurrentSynapse(current=60*float(z)))

E1 = electrode_names['AVBL']
E2 = electrode_names['AVBR']
E3 = electrode_names['PVCR']
E4 = electrode_names['PVCL']
E5 = electrode_names['AVAL']
E6 = electrode_names['AVAR']
E7 = electrode_names['AVDL']
E8 = electrode_names['AVDR']

for k in dorsal_muscle_electrodes_dict:
    record = []
    for i in range(duration):
        B.step()
        E1.pin(10)
        E2.pin(10)
        E3.pin(10)
        E4.pin(10)
        E5.pin(10)
        E6.pin(10)
        E7.pin(10)
        E8.pin(10)

        record.append(dorsal_muscle_electrodes_dict[k].measure())

    rec.append(record)
    

'''
for u,v in connectome.edges():
    B.add_synapse(u, v, BasicCurrentSynapse(current=60))


E1 = electrode_names['ADAL']
E2 = electrodes[neuron_names['MDR07']]
E3 = electrodes[neuron_names['MDL07']]

for i in range(duration):
    B.step()
    E1.pin(2)
    E2.pin(2)
    E3.pin(2)
    
    rec1.append(E3.measure())
    
    rec.append((E1.measure(),E2.measure(),E3.measure()))

Need to understand what pin does.

Goals:
Replicate Harvard study. Simulate command interneurons.
    We have a brain that propogates based on inhibitory/excitatory synapses and weights. 
    We have inhibitory/excitatory synapses implemented. Our connectome has negative weights. 
    DDs to muscle nodes have negative weights (GABAergic synapses). AWESOME.
    VBs/VDs feed to DDs (excitatory, Ach). 
Include ablation code and connectome construction code. 
We want to put electrodes on a group specified neurons.
Make sure weights are normalized.
'''
```
